#可选择参数

##数据参数
*   mode train
*   template 
*   test_model_path
*   dataset_code default m1-20m
*   min_rating
*   min_uc keepuser
*   min_sc keep item
*   split leave_one_out
*   dataloader_code bert
*   dataloader_random_seed
*   train_batch_size
*   train_negative_sampler_code random
*   train_negative_size
*   test_negative_sampling_seed


##训练参数
*   trainer_code
*   device
*   num_gpu
*   optimizer Adam or SGD
*   lr 0.001
*   weight_decay l2 regularization\
*   momentum SGD
*   decay_step for StepLR
*   gamma for StepLR
*   num_epochs
*   log_period_as_iter
*   metric_ks ks@k[10,20,50]
*   best_metric NDCG@10 desicion for best model

##模型参数
*   bert_max_len   
*   bert_num_items  (vocab_size_)
*   bert_hidden_units hidden_size
*   bert_num_blocks  nums of layers
*   bert_num_heads  nums of head for MA
*   bert_dropout 
*   bert_mask_prob
